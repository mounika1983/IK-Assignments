Data Cleaning

````python
dict_2 = {
    'id': [1,2,np.nan,np.nan,5],
    'name': [np.nan,'B','Matt','A','D'],
    'age': [15, 12, np.nan, np.nan, 22]
    }
````

````python
df2 = pd.DataFrame(data = dict_2, index = ['Person1','Person2','Person3','Person4','Person5'])
print(df2)
````

````python
df2.isnull().sum()/len(df2)
````

````python
len(df2)
````

````python
# identifying missing data
missing_data = df2.isnull()
print(missing_data)
````

````python
not_missing_data = df2.notnull()
print(not_missing_data)
````

````python
df2.fillna(0)
````

````python
# filling missing data
df2.fillna(value = 0)
````

````python
print(df2)
````

````python
# dropping rows or columns with missing data
df2.dropna(axis = 0) # dropping rows with missing data
````

````python
df2.dropna(axis = 1) # dropping columns with missing data
````

Data Aggregation

````python
print(df)
````

````python
df.groupby('name')['age'].mean()
````

````python
grouped_df = df.groupby('name')
print(grouped_df)
````

````python
df.groupby('name').sum()
````

````python
def cust_func(y:pd.Series):
  avg = y.sum()/len(y)
  return avg
````

````python
print(df)
````

````python
df.groupby('name').agg(cust_func)
````

````python
df.groupby('name').agg({'id': sum, 'age': np.mean})
````

````python
groupby = df.groupby('name').agg({'id': sum, 'age': np.mean})
````

````python
groupby.reset_index()
````
